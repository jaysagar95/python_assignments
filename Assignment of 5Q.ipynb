Q5 Write a Python program to push all zeros to the end of a list. Input : [0,2,3,4,6,7,10]

```python
list = []
while True:
  a=input('enter the number ')
  if a == '':
    break
  a=int(a)  
  list.append(a)
print(list)
for i in list:
  if i==0:
    list.remove(i)
    list.append(i)
print(list)    

```
